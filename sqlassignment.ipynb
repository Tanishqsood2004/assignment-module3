Purpose of Constraints in a Database:

Constraints are rules applied to database tables to ensure the accuracy, reliability, and integrity of the data. They define specific conditions that the data must meet to be inserted or updated in the database.

 Constraints help:

Maintain Data Integrity: Ensure that data stored in the database is valid and consistent.

Enforce Business Rules: Prevent invalid data entry by enforcing rules such as unique values, non-null fields, or valid relationships.

Improve Database Reliability: Reduce errors and anomalies by automatically rejecting data that violates constraints.

# Q 4 3.Why would you apply the NOT NULL constraint to a column? Can a primary key contain NULL values? Justify your answer.

Why Apply the NOT NULL Constraint to a Column?

The NOT NULL constraint ensures that a column cannot contain NULL values. This is used in scenarios where a value is always required and cannot be left unspecified.

Reasons to Apply NOT NULL:

Enforce Mandatory Data: For example, critical fields like employee_id or email must always have a value.

Prevent Incomplete Records: Ensures no essential data is missing when inserting or updating rows.

Support Business Logic: Reflects real-world requirements where certain attributes (like a product price or customer name) cannot logically be absent.

Can a Primary Key Contain NULL Values?
No, a primary key cannot contain NULL values. This is because a primary key is used to uniquely identify each row in a table, and NULL represents an unknown or undefined value, which violates the uniqueness requirement.

Justification:
Primary Key Requirements:

A primary key must uniquely identify each record in a table.
All values in the primary key column(s) must be unique and non-NULL.
Why NULL Is Not Allowed:

NULL is not a value but a marker for missing or undefined data. Allowing NULL in a primary key would create ambiguity in uniquely identifying rows.
For example, if a primary key contains NULL, the database cannot determine whether it represents a unique value or an unknown one.

# Q 5 5. Explain the consequences of attempting to insert, update, or delete data in a way that violates constraints. Provide an example of an error message that might occur when violating a constraint

Consequences of Violating Constraints
When attempting to insert, update, or delete data that violates a constraint, the database management system (DBMS) enforces the rule and rejects the operation. This ensures that the integrity of the data is maintained.

The consequences are:

Insertion Failure: An INSERT statement that violates a constraint is rolled back, and no new row is added.
Update Failure: An UPDATE statement that violates a constraint is rolled back, leaving the data unchanged.
Deletion Failure: A DELETE statement that violates referential integrity (e.g., due to a foreign key constraint) is rolled back, and no rows are deleted.
