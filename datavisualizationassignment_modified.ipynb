5. Show a pie chart to represent the percentage distribution of different sections in the array `sections`.sections = ['Section A', 'Section B', 'Section C', 'Section D']

sizes = [25, 30, 15, 30]

```python
import matplotlib.pyplot as plt

# Given data
sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]

# Create pie chart
plt.pie(sizes, labels=sections, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'yellowgreen', 'gold'])

# Adding title
plt.title('Percentage Distribution of Sections')

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 100

# Generate correlated features
mean = [0, 0, 0]
cov = [[1, 0.8, 0.5],  # Features are correlated
       [0.8, 1, 0.3],
       [0.5, 0.3, 1]]

# Create synthetic dataset
data = np.random.multivariate_normal(mean, cov, n_samples)
columns = ['Feature1', 'Feature2', 'Feature3']
df = pd.DataFrame(data, columns=columns)

# Display first few rows of the dataset
print(df.head())
```
