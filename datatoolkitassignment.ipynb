# Q1 Demonstrate three different methods for creating identical 2D arrays in NumPy Provide the code for each

Here are three different methods to create identical 2D arrays in NumPy:

Method 1: Using np.array()
You can directly use np.array() to create an array by specifying its values.

```python
import numpy as np

# Method 1: Using np.array()
array1 = np.array([[1, 2, 3], [4, 5, 6]])

print("Array 1:")
print(array1)
```

Method 2: Using np.zeros() and then assigning values
You can create a zero-initialized array and then assign values to it.

```python
# Method 2: Using np.zeros() and then assigning values
array2 = np.zeros((2, 3))
array2[0, :] = [1, 2, 3]
array2[1, :] = [4, 5, 6]

print("\nArray 2:")
print(array2)
```

Method 3: Using np.full()
You can use np.full() to create an array filled with a specific value, and then update individual elements.

```python
# Method 3: Using np.full() and updating values
array3 = np.full((2, 3), 0)
array3[0, :] = [1, 2, 3]
array3[1, :] = [4, 5, 6]

print("\nArray 3:")
print(array3)
```

# Q 3  Explain the following termsg
M The difference in npYarray, npYasarray and npYasanyarrayX
M The difference between Deep copy and shallow copyX

1. Difference Between np.array(), np.asarray(), and np.asanyarray()
np.array():

This function always creates a new array, even if the input is already an array. It copies the input data into a new NumPy array by default unless specified otherwise.
Example

```python
import numpy as np
lst = [1, 2, 3]
arr = np.array(lst)
```

Key Point: It creates a new array and allows for changes in data type, size, or memory allocation.
np.asarray():

This function converts the input into an array only if it's not already a NumPy array. If the input is already a NumPy array, it returns the input unchanged (i.e., no new copy is made). It’s used to avoid unnecessary copying when working with NumPy arrays.
Example

```python
arr1 = np.array([1, 2, 3])
arr2 = np.asarray(arr1)  # Returns the same array, no copy made
```

Key Point: It is faster when the input is already an array, as it avoids copying the data.
np.asanyarray():

Similar to np.asarray(), but it allows subclasses of ndarray (such as np.matrix) to pass through unchanged. np.asarray() will always convert subclasses into a base ndarray, while np.asanyarray() will preserve the subclass.


```python
arr1 = np.matrix([[1, 2], [3, 4]])
arr2 = np.asanyarray(arr1)  # Returns a matrix, not a base ndarray
```

Key Point: It preserves the subclass type (like matrix) rather than always returning a base ndarray.
2. Difference Between Deep Copy and Shallow Copy
Shallow Copy:

A shallow copy creates a new object, but does not copy the nested objects. It simply references the original nested objects. Changes to mutable nested objects will be reflected in both the original and the copied object.
In NumPy, shallow copy can be done using view().


```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = arr1.view()  # Shallow copy
arr2[0] = 100
print(arr1)  # [100  2  3], changes affect original array
```

Deep Copy:

A deep copy creates a completely independent object, including its nested objects. Changes to any part of the new object will not affect the original object.
In NumPy, deep copy can be done using copy().


```python
arr1 = np.array([1, 2, 3])
arr2 = arr1.copy()  # Deep copy
arr2[0] = 100
print(arr1)  # [1 2 3], original array remains unchanged
```

Summary:
np.array() always creates a new array, np.asarray() avoids copying if the input is already an array, and np.asanyarray() retains subclasses like matrix.
Shallow copy shares references to nested objects, whereas deep copy duplicates everything, ensuring changes are independent.

# Q 6 ( Create a D NumPy array of shape (, , ) containing random integers Petween w and wR Perform the
following operations:

 a) Find the indices of the maximum values along each depth level (third axis).

 b) Perform elementwise multiplication of between both arrayX

```python
import numpy as np

# Generate a 3x3x3 array with random integers between 1 and 10
array = np.random.randint(1, 11, (3, 3, 3))

print("Original 3D Array:")
print(array)

# a) Find the indices of the maximum values along the depth level (third axis)
max_indices = np.argmax(array, axis=2)

print("\nIndices of Maximum Values Along Each Depth Level:")
print(max_indices)

# b) Perform element-wise multiplication between both array (the array and max_indices)
# Since max_indices is 2D, we need to adjust the dimensions before multiplying
# Here we broadcast max_indices to the 3D shape of 'array' to perform multiplication
# Broadcasting is done by adding a new axis

broadcasted_max_indices = np.expand_dims(max_indices, axis=2)
result = array * broadcasted_max_indices

print("\nElement-wise Multiplication Between Array and Broadcasted Indices:")
print(result)
```

Explanation:
np.random.randint(1, 11, (3, 3, 3)) generates a random 3D array with integers between 1 and 10.
np.argmax(array, axis=2) finds the indices of the maximum values along the third axis (depth level).
np.expand_dims(max_indices, axis=2) reshapes max_indices so that it can be broadcasted for element-wise multiplication with the original 3D array.

To clean and transform the 'Phone' column in the dataset, we will remove non-numeric characters and convert the column to a numeric data type. Here's how you can do this using Python and pandas:

Steps:
Remove non-numeric characters: Use a regular expression to remove any character that is not a digit.
Convert to numeric: Convert the cleaned column to a numeric data type.
Display the table attributes and data types: Use pandas' dtypes to show the data types of each column.
Code:

```python
import pandas as pd

# Load your dataset (assuming the dataset is already loaded as 'df')
# If your dataset is not loaded, you can use pd.read_csv() or similar methods to load it.

# Sample data for demonstration
data = {'Name': ['John Doe', 'Jane Smith', 'Alice Johnson'],
        'Phone': ['(123) 456-7890', '123.456.7890', '+1-123-456-7890']}

df = pd.DataFrame(data)

# Clean and transform the 'Phone' column to remove non-numeric characters
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True)

# Convert 'Phone' column to numeric data type (int64)
df['Phone'] = pd.to_numeric(df['Phone'])

# Display the cleaned DataFrame
print("Cleaned DataFrame:")
print(df)

# Display table attributes and data types of each column
print("\nTable attributes and data types:")
print(df.dtypes)
```

Explanation:
str.replace(r'\D', '', regex=True): This regular expression replaces all non-digit characters (\D) with an empty string.
pd.to_numeric(df['Phone']): Converts the Phone column to a numeric data type (int64).
df.dtypes: Displays the data types of all columns in the DataFrame.

# Q 17 Kçg Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays (x and y)
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 10 - 5  # Random values between -5 and 5
y = np.random.rand(100) * 10 - 5  # Random values between -5 and 5

# Create a DataFrame for the scatter plot
data = pd.DataFrame({'X': x, 'Y': y})

# Determine the quadrant for each point
def get_quadrant(row):
    if row['X'] >= 0 and row['Y'] >= 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] >= 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', style='Quadrant', palette='Set1', s=100)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Add a legend
plt.legend(title='Quadrants')

# Show the plot
plt.axhline(0, color='black', lw=0.5, ls='--')  # X-axis
plt.axvline(0, color='black', lw=0.5, ls='--')  # Y-axis
plt.grid()
plt.show()
```

Explanation of the Code:
Generate Random Data:

Two random arrays x and y are generated with values ranging from -5 to 5 using np.random.rand().
A DataFrame is created to hold these arrays for easier manipulation and plotting.
Determine Quadrants:

A function get_quadrant() is defined to determine the quadrant for each point based on its x and y coordinates.
The quadrant information is added to the DataFrame as a new column.
Seaborn Scatter Plot:

A scatter plot is created using Seaborn’s scatterplot() function, coloring points based on their quadrants.
The hue parameter is used to color the points according to the quadrant they belong to, and style can help differentiate them in the legend.
Axes Labels and Title:

The x-axis and y-axis are labeled, and the title of the plot is set to "Quadrant-wise Scatter Plot".
Legend:

A legend is added to the plot to indicate which colors correspond to which quadrants.
Grid and Axes:

Dashed lines are added at the x=0 and y=0 positions to indicate the axes./

# Q 18 *58 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'8

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine values for each x

# Create a Bokeh plot
output_notebook()  # Use this if you're running in a Jupyter Notebook
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(X)', line_width=2, color='blue')

# Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

# Q 19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Cateorical
Bar Chart'8

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Prepare random categorical data
categories = [f'Category {i+1}' for i in range(10)]  # 10 categories
values = np.random.randint(10, 100, size=len(categories))  # Random values for each category

# Create a DataFrame for better handling of data
data = pd.DataFrame({'Categories': categories, 'Values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a Bokeh bar chart
output_notebook()  # Use this if you're running in a Jupyter Notebook
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

# Map colors based on values
mapper = linear_cmap('Values', Viridis256, low=min(values), high=max(values))

# Add bars to the chart with color mapping
bars = p.vbar(x='Categories', top='Values', source=source,
              width=0.9, line_color='white', fill_color=mapper)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Categories"), ("Value", "@Values")]
p.add_tools(hover)

# Label the axes
p.xgrid.grid_line_color = None  # Hide grid lines on the x-axis
p.y_range.start = 0  # Start y-axis from 0

# Show color bar (optional)
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```

Explanation of the Code:
Import Libraries:

Necessary libraries from Bokeh for plotting, displaying, and handling data are imported.
Prepare Random Categorical Data:

A list of categories is generated (10 categories).
Random values between 10 and 100 are generated for each category using np.random.randint().
Create a DataFrame:

A Pandas DataFrame is created for easier handling of the data.
Create a ColumnDataSource:

The ColumnDataSource is created from the DataFrame, which allows for efficient data sharing between Bokeh plots and data.
Create a Bokeh Bar Chart:

The figure() function initializes the plot, setting the range for the x-axis based on categories, along with titles and labels for the axes.
Add Bars to the Chart:

The vbar() method adds vertical bars to the chart based on the values in the ColumnDataSource.
Bars are colored based on their values.
Add Hover Tooltips:

The HoverTool is configured to show the category and value when hovering over the bars.
Label the Axes:

The x-grid lines are hidden for a cleaner look, and the y-axis starts from 0.
The legend's location is set.
Show the Plot:

Finally, show(p) displays the bar chart.

# Q 2 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'8


```python
import plotly.graph_objs as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # 100 random values between 0 and 1

# Create a line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    template='plotly_white'  # Optional: gives a clean background
)

# Show the plot
fig.show()
```

Explanation of the Code:
Import Libraries:

Import plotly.graph_objs for creating the plot and numpy for generating random data.
Generate Random Data:

np.linspace(0, 10, 100) creates an array of 100 equally spaced values from 0 to 10 for the x-axis.
np.random.rand(100) generates 100 random values between 0 and 1 for the y-axis.
Create a Line Plot:

A Figure object is created using go.Figure().
The add_trace() method adds a line plot (go.Scatter()) to the figure. The mode='lines' argument specifies that we want to create a line plot.
Update Layout:

The update_layout() method is used to set the title and axis labels. You can also apply a template for styling (e.g., 'plotly_white').
Show the Plot:

Finally, fig.show() displays the plot.

# Q 21 Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import plotly.graph_objs as go
import numpy as np

# Generate random data
labels = [f'Category {i+1}' for i in range(5)]  # 5 categories
values = np.random.randint(10, 100, size=len(labels))  # Random values for each category

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='label+percent',
                               insidetextorientation='radial')])

# Update layout
fig.update_layout(
    title='Interactive Pie Chart',
    template='plotly_white'  # Optional: gives a clean background
)

# Show the plot
fig.show()
```

Explanation of the Code:
Import Libraries:

Import plotly.graph_objs for creating the pie chart and numpy for generating random data.
Generate Random Data:

labels creates a list of 5 category names.
values generates random integers between 10 and 100 for each category.
Create a Pie Chart:

A Figure object is created with a go.Pie() trace. The textinfo='label+percent' argument specifies that both the labels and the percentages should be displayed on the pie chart. The insidetextorientation='radial' argument ensures that the text is oriented correctly.
Update Layout:

The update_layout() method is used to set the title of the pie chart. You can also apply a template for styling (e.g., 'plotly_white').
Show the Plot:

Finally, fig.show() displays the interactive pie chart.
