**Question 11:**   Identify and remove outliers in the km_driven column using the IQR method?

**Question 10:** How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
violin plot.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/drive/MyDrive/spotify.csv'
spotify_data = pd.read_csv(file_path)

# Filter the data for artists with multiple tracks to keep the plot readable
top_artists = spotify_data['Artist'].value_counts().head(10).index
filtered_data = spotify_data[spotify_data['Artist'].isin(top_artists)]

# Create the violin plot
plt.figure(figsize=(14, 8))
sns.violinplot(data=filtered_data, x='Artist', y='Popularity', palette='muted')

# Customize the plot
plt.title('Distribution of Track Popularity for Top Artists', fontsize=16)
plt.xlabel('Artist', fontsize=14)
plt.ylabel('Popularity', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
